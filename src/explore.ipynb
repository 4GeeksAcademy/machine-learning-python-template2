```python
import pandas as pd

df = pd.read_csv('/workspaces/machine-learning-python-template2/data/raw/bank-marketing-campaign-data.csv', sep = ';')
df
```

```python
df.drop(["contact", "month", "day_of_week", "duration", 'campaign', 'pdays'], axis = 1, inplace = True)
df.head()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns

fig, axis = plt.subplots(4, 2, figsize = (15, 10.5))


cols = ['job', 'marital', 'education', 'default', 'housing', 'poutcome', 'y']

for col, ax in zip(cols, axis.ravel()): 
    sns.histplot(ax = ax, data = df, x = col, color = 'darkviolet').set(ylabel = None)



plt.tight_layout()


plt.show()
```

```python
fig, axis = plt.subplots(4, 2, figsize = (15, 10.5))


cols = ['job', 'marital', 'education', 'default', 'housing', 'poutcome', 'y']

for col, ax in zip(cols, axis.ravel()): 
    sns.countplot(ax = ax, data = df, x = 'y', hue =col)



plt.tight_layout()
```

```python
fig, axis = plt.subplots(2, 6, figsize = (10, 7))


x_data = ['age', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']

for i, col in enumerate(x_data):
    sns.histplot(ax=axis[0, i], data=df, x=col, color='darkviolet').set(ylabel=None)
    sns.boxplot(ax=axis[1, i], data=df, x=col, color='darkviolet')




plt.show()
```

```python
fig, axis = plt.subplots(2, 6, figsize = (10, 7))
df["y_n"] = pd.factorize(df["y"])[0]

x_data = ['age', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']

for i, col in enumerate(x_data):
    sns.regplot(ax = axis[0, i], data = df, x = col, y = 'y_n', color = 'darkviolet')
    sns.heatmap(df[["y_n", col]].corr(), annot = True, fmt = ".2f", ax = axis[1, i], cbar = False)




plt.show()
```
